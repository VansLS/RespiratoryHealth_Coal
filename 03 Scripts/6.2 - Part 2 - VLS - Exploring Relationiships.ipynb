```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import os
```

```python
df=pd.read_csv(os.path.join(path,'02 Data','Prepared Data','WHO_Data.csv'))
```

```python
df.head()
```

```python
# Create a categorical plot in seaborn using the Resp Deaths % levels created 
sns.set(style="ticks")
g = sns.catplot(x="Age Group Flag", y="Resp Deaths %", hue="Resp Deaths % levels", data=df)
plt.xticks(rotation=45)
plt.show()
```

## There are some 'high levels' values that are much higher than the rest. This needs further investigation.

```python
sns.barplot(x='Region Name', y='Resp Deaths %', data=df.sort_values(by='Resp Deaths %'))
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 6)  # Set y-axis limits from 0 to 6
plt.yticks(range(0, 7, 1))  # Set y-axis tick intervals from 0 to 6 with step of 1
plt.show()
```

## Europe has the least % of respiratory deaths. Asia, Africa and Oceania have the highest respiratory death %. 

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='Year', y='Resp Deaths %', data=df.sort_values(by='Year'))
plt.xticks(rotation=80, ha='right')
plt.show()
```
