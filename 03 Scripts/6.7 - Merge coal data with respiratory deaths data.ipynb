```python
import pandas as pd
import numpy as np
import os
```

```python
path = r'/Users/vanessalimsingh/Achievement 6 June 2024'
```

```python
whodf= pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data' , 'who_df_v2_updatedregionnames.csv'))
```

```python
whodf.head()
```

```python
whodf_2 = whodf[(whodf['Age Group'] == '[All]') & (whodf['Sex'] == 'All')]
```

```python
whodf_2.head()
```

```python
#create dataset with unique years, 'all' sex and 'all age group'
whodf_2 = whodf_2.drop(['Unnamed: 0', 'AS_Death rate/100 000', 'Death rate/100 000','Age Group Flag'], axis=1)
```

```python
whodf_2.head()
```

```python
whodf_2.duplicated().sum() #no dupes. 
```

```python
coal= pd.read_csv(os.path.join(path, '02 Data', 'Original Data' , 'coal-per-capita-co2.csv'))
```

```python
coal=coal.rename(columns={'Entity':'Country'})
```

```python
coal.head()
```

```python
subset_coal = coal[(coal['Year'] >= 1950) & (coal['Year'] <= 2022)]
```

```python
subset_coal.head()
```

```python
subset_coal.duplicated().sum() #no dupes
```

```python
#merge subset_coal and whodf_2
```

```python
merged_df = pd.merge(subset_coal, whodf_2, on=['Country', 'Year'])
```

```python
merged_df.head()
```

```python
merged_df.duplicated().sum() #no dupes
```

```python
#export data for analysis
```
